```python
events = np.column_stack((np.array(event_timestamp, dtype = int),
                        np.zeros(len(event_timestamp), dtype = int),
                        np.array(event_index, dtype = int)))
fre1 = 8
fre2 = 10
fre3 = 12
fre4 = 14
events_id = { # Set up your event name
     f'natural': 0,
     f'Frequency {fre1} Hz': 1,
     f'Frequency {fre2} Hz': 2,
     f'Frequency {fre3} Hz': 3,
     f'Frequency {fre4} Hz': 4,
}
# events_id = { # Set up your event name
#      f'Frequency {fre1} Hz': 1,
#      f'Frequency {fre2} Hz': 2,
#      f'Frequency {fre3} Hz': 3,
#      f'Frequency {fre4} Hz': 4,
# }

mne_epochs = mne.Epochs(raw_mne, events, 
        tmin= -0.5,     # init timestamp of epoch (0 means trigger timestamp same as event start)
        tmax= 10,    # final timestamp (10 means set epoch duration 10 second)
        event_id =events_id,
        preload = True,
        event_repeated='drop',
        baseline=(-0.5, 0)
    )
```

```python
# Get EEG data and events
X = mne_epochs.get_data()
y = mne_epochs.events[:, -1]
print(y.shape)
print(X.shape)
import numpy as np

# สร้างตัวแปรเก็บข้อมูลที่เหลือ
X_filtered = []
y_filtered = []

count_class_0 = 0
max_class_0 = 20

for i in range(len(y)):
    if y[i] == 0 and count_class_0 < max_class_0:
        # เพิ่มข้อมูล class 0 และเพิ่มจำนวน class 0 ที่เหลือ
        X_filtered.append(X[i])
        y_filtered.append(y[i])
        count_class_0 += 1
    elif y[i] != 0:
        # เพิ่มข้อมูล class อื่นๆ ทั้งหมด
        X_filtered.append(X[i])
        y_filtered.append(y[i])

# แปลงเป็น numpy array
X_filtered = np.array(X_filtered)
y_filtered = np.array(y_filtered)

# Initilize CSP
csp = CSP(n_components = 8, norm_trace = False)

# Fit CSP to data 
csp.fit(X_filtered, y_filtered)

# Visualize CSP patterns
# csp.plot_patterns(mne_epochs.info)
```

```python
# Transform data into CSP space
X_transformed = csp.transform(X_filtered)
# print(X_transformed.shape)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y_filtered, test_size = 0.2, random_state = 42)
# print(X_test[0])
# print(X_test[1])
# print(X_test)
# print(X_train.shape)
# print(X_test.shape)
# # print(y_test.shape)
# # print(y_train.shape)
# print(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC 
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

# Create and fit the logistic regression model within a pipeline
clf = Pipeline([('LR', LogisticRegression())])
clf.fit(X_train, y_train)

# Predict and calculate accuracy
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
array_con = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy}")

# report = classification_report(y_test, y_pred, target_names=[f'Frequency {fre1} Hz', f'Frequency {fre2} Hz', f'Frequency {fre3} Hz', f'Frequency {fre4} Hz'])
report = classification_report(y_test, y_pred, target_names=['natural', f'Frequency {fre1} Hz', f'Frequency {fre2} Hz', f'Frequency {fre3} Hz', f'Frequency {fre4} Hz'])  
print(report)
print(array_con)
```

```python
from sklearn.svm import SVC 

clf = Pipeline([('SVM', SVC(kernel='linear'))])  
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

report = classification_report(y_test, y_pred, target_names=['natural', f'Frequency {fre1} Hz', f'Frequency {fre2} Hz', f'Frequency {fre3} Hz', f'Frequency {fre4} Hz'])  # adjust target names accordingly
print(report)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

clf = Pipeline([('LDA', LDA())])
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

# print(len(X_test))
# print(X_train)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

report = classification_report(y_test, y_pred, target_names=['natural', f'Frequency {fre1} Hz', f'Frequency {fre2} Hz', f'Frequency {fre3} Hz', f'Frequency {fre4} Hz']) 
print(report)
```

```python
# CSP-LDA pipeline

csp = CSP(n_components=4, log=True, norm_trace=False)
lda = LDA()

pipeline = Pipeline([('CSP', csp), ('LDA', lda)])

scores = cross_val_score(pipeline, X_filtered, y_filtered, cv=5)  # 5-fold cross-validation
print(scores)
```
